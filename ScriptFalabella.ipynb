```python
import yfinance as yf

# Incorporando el ticker de Falabella
Falabella_ticker = yf.Ticker('FALABELLA.SN')

# Para recoger la historia
Falabella_df = Falabella_ticker.history(start='2020-01-01')
```

```python
import plotly.express as px

# Para generar el gráfico de lineas
fig = px.line(x=Falabella_df.index, y=Falabella_df['Close'])

# Actualización del layout
fig.update_layout(
    title='Precios de Cierre de Falabella',
    xaxis_title='Fechas',
    yaxis_title='Precio',
    template='plotly_white',
    xaxis_rangeslider_visible=True
)

# Para mostrar el gráfico
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Creating figure with second y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Adding line plot with close prices and bar plot with trading volume
fig.add_trace(go.Scatter(x=Falabella_df.index, y=Falabella_df['Close'], name='Cierre Falabella'), secondary_y=False)
fig.add_trace(go.Bar(x=Falabella_df.index, y=Falabella_df['Volume'], name='Volumen', opacity=0.5), secondary_y=True)

# Updating layout
fig.update_layout(
    title='Precios de Cierre de Falabella',
    xaxis_title='Fechas',
    yaxis_title='Precio',
    template='plotly_white',
    xaxis_rangeslider_visible=True,
    hovermode='x'
)

# Disabling grid of second y-axis
fig.layout.yaxis2.showgrid=False

# Showing figure
fig.show()
```
