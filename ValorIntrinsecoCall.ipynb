```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['font.family'] = 'serif'#serif es un ejemplo de tipografía
```

```python
# Precio de Ejercicio de la Opción:
K = 1000
# Gráfico
S = np.linspace(800, 1200, 1000) # Valores del dólar
h = np.maximum(S - K, 0) # Valores intrínsecos de la opción call
plt.figure() #Esta instrucción crea una nueva figura
plt.plot(S, h, lw=2.5) # dibujar valores intrínsecos al vencimiento
plt.xlabel('Nivel del dólar $S_t$ al vencimiento') #Nombre etiqueta horizontal
plt.ylabel('Valor intrínseco de opción call') #Nombre etiqueta vertical
plt.grid(True)
```
